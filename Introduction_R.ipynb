## Manual Mean calculation function (Without in-built function)

```r
mean_manual <- function(x)
{
  l  <-  length(x)
  sum <- 0
  for (i in 1:l)
  {
    sum <- sum + x[i]
  }
  return(sum/l)
}
```

```r
variance_manual=function(x)
{
    sum <- 0
    l <- length(x)
    for (i in 1:l)
    {
    sum  <-   sum+((x[i]-mean_manual(x))^2)
    }
    return(sum/(l-1))
}
```

```r
std_manual=function(x){
    return((variance_manual(x))^0.5)
}
```

## Creating function for Euclidean distance

```r
eucl_dist_manual <- function(x1, x2)
{
  sqrt(sum((x1 - x2) ^ 2))  
} 
```

## Calculating and sorting in ascending order the eucidian dist of created sample and each row in iris dataset

```r
?apply()
```

```r
distances <- apply(iris_nospecies,1,function(x) eucl_dist_manual(x,custom_vector))
distances <- sort(distances)
distances
```
